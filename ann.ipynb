```python
import pandas as pd
import numpy as np
```

```python
import keras,tensorflow,scikeras
```

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout
```

```python
df=pd.read_csv(r'cleancreditcarddata.csv')
pd.set_option('display.max_columns',None)
df
```

```python
y=df['Attrition_Flag']
x=df.drop('Attrition_Flag',axis=1)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,roc_auc_score
from sklearn.model_selection import train_test_split,cross_val_score
from scikeras.wrappers import KerasClassifier
```

```python
#splitting data
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=22)
```

```python
from imblearn.over_sampling import SMOTE
us=SMOTE()
x_train,y_train=us.fit_resample(x_train,y_train)
```

```python
x_train.shape,y_train.shape,
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x_train=ss.fit_transform(x_train)
x_test=ss.transform(x_test)
```

```python
classifier=Sequential()
```

```python
classifier.add(Dense(input_dim=23,units=46,kernel_initializer='uniform',activation='relu'))
classifier.add(Dropout(0.2))
classifier.add(Dense(units=1,activation='sigmoid'))
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
classifier.fit(x_train,y_train,epochs=20,batch_size=32)
```

```python
ypred_train=classifier.predict(x_train)
ypred_train=(ypred_train>0.5).astype(int)
ypred_test=classifier.predict(x_test)
ypred_test=(ypred_test>0.5).astype(int)
```

```python
print('train:',accuracy_score(y_train,ypred_train))
print('test:',accuracy_score(y_test,ypred_test))
```

```python
# cross validation
```

```python
def classifier_model():
    ann_classifier=Sequential()
    ann_classifier.add(Dense(input_dim=23,units=46,kernel_initializer='uniform',activation='relu'))
    ann_classifier.add(Dropout(0.2))
    ann_classifier.add(Dense(units=1,activation='sigmoid'))
    ann_classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
    return ann_classifier
```

```python
cv=KerasClassifier(classifier_model,batch_size=32,epochs=25)
```

```python
print('cv:',cross_val_score(cv,x_train,y_train,cv=5,scoring='accuracy').mean())
```
