```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,roc_auc_score
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.ensemble import RandomForestClassifier
```

#saving the model

```python
dt=RandomForestClassifier()
```

```python
# spliting data
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=22)
```

```python
x_train.head()
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
y_train.value_counts()
```

```python
#treating imablance data
```

```python
from imblearn.over_sampling import SMOTE
us=SMOTE()
x_train,y_train=us.fit_resample(x_train,y_train)
```

```python
x_train.shape,y_train.shape,
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
estimator=RandomForestClassifier()
param_grid={'n_estimators':list(range(1,50))}
grid=GridSearchCV(estimator,param_grid,cv=5,scoring='accuracy')
grid.fit(x_train,y_train)
grid.best_params_
```

```python
rfc=grid.best_estimator_
rfc
```

```python
#extractin  imporatant columns
```

```python
feat_imp=pd.DataFrame(rfc.feature_importances_,index=x_train.columns,columns=['importances'])
feat_imp
```

```python
imp=feat_imp[feat_imp['importances']>0].index.tolist()
imp
```

```python
type(imp)
```

```python
x_train=x_train[imp]
x_test=x_test[imp]
```

```python

ypred_train=rfc.predict(x_train)
print('train:',accuracy_score(y_train,ypred_train))
print('cv:',cross_val_score(rfc,x_train,y_train,cv=5,scoring='accuracy').mean())
```

```python

ypred_train=rfc.predict_proba(x_train)[:,1]
plt.hist(ypred_train)
plt.show()
```

```python
ypred_test=rfc.predict(x_test)
print('test:',accuracy_score(y_test,ypred_test))
```

```python
# classificatio report
```

```python
print(classification_report(y_test,ypred_test))
```

```python
print(roc_auc_score(y_train,ypred_train))
```
